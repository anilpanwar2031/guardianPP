```python
import pdfplumber

file_path = "SZN67RA4OH.pdf"
```

```python
keys = ['Procedure', 'DeltaDentalPPO', 'DeltadentalPremier','Nonpardentist','Frequency','Agelimitation','LimitationORNotes']
```

```python
tab = []
with pdfplumber.open(file_path) as pdf:
  
    for page in pdf.pages:
        tables = page.extract_tables()
        for i , table in enumerate(tables):
#             print("\n")
#             print("Table", table)
#             print("Table", len(table))
#             print("\n")
            for r in table:
                td = {}
                if len(r) == 7:
                    td['Procedure'] = r[0]
                    td['DeltaDentalPPO'] = r[1]
                    td['DeltadentalPremier'] = r[2]
                    td['Nonpardentist'] = r[3]
                    td['Frequency'] = r[4]
                    td['Agelimitation'] = r[5]
                    td['LimitationORNotes'] = r[6]
                    tab.append(td)
                    
               
               
                    
```

```python
tab_list = []
index = []
for i, t in enumerate(tab):
    if "Procedure" in t['Procedure']:
        index.append(i+1)
    if "Individual Annual\nDeductible Met\nPPO Network" in t['DeltaDentalPPO']:
        index.append(i-1)
        
```
